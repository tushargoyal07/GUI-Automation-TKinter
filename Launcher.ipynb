```python
import os
import tkinter as tk
from tkinter import *
import tkinter as tk
from tkinter import *
from tkinter import ttk
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,r2_score
from sklearn.impute import SimpleImputer
from tkinter.filedialog import askopenfilename
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import subprocess
import os
import nbformat
from nbconvert.preprocessors import ExecutePreprocessor


root= Tk()
root.title('Launcher')
root.geometry('800x450')
#root.configure(fg=)

def run_notebook(filepath):
    # Load the notebook
    with open(filepath, "r") as f:
        nb = nbformat.read(f, as_version=4)

    # Execute the notebook
    ep = ExecutePreprocessor(timeout=None)
    ep.preprocess(nb, {"metadata": {"path": os.path.dirname(filepath)}})

    # Save the executed notebook
    with open(filepath, "w", encoding="utf-8") as f:
        nbformat.write(nb, f)


def Kmeans():
    filepath = r"C:\Users\WELCOME\Desktop\Minor_Project\Kmeans.ipynb"
    run_notebook(filepath)
    


def LinReg():
    filepath = r"C:\Users\WELCOME\Desktop\Minor_Project\LinearReg.ipynb"
    run_notebook(filepath)



Label(root,text="Choose the model based on the type of problem",font="System").place(x=230,y=20) 


Label(root,text="Regression",font="System").place(x=150,y=75)
Label(root,text="Clustering",font="System").place(x=500,y=75)



Button(root,text='K-Means',command=Kmeans,activebackground="black",activeforeground="white").place(x=500,y=130)

Button(root,text='Linear Regression',activebackground="black",command=LinReg,activeforeground="white").place(x=150,y=130)


root.mainloop()
```
